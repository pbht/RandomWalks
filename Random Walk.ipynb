```python
# We simulate random walks
```

```python
import random as rand
import matplotlib.pyplot as plt
```

```python
def walk(steps, bias):
    walk = [0]
    cumulative = 0
    for step in steps:
        cumulative += step
        cumulative += bias
        walk.append(cumulative)
    
    return walk
```

```python
def plot_walks(n, bias=0):
    steps = [-1, 1]
    x_axis = [i for i in range(0, n+1)]

    for _ in range(n):
        step_history = []
        for _ in range(n):
            step_history.append(rand.choice(steps))
        walk_list = walk(step_history, bias)   
        plt.plot(x_axis, walk_list)
   
    plt.plot(x_axis, [bias*x for x in x_axis], color='black', label='Bias')

    plt.legend()
    plt.show()
```

```python
plot_walks(100, bias=-0.1)
```
